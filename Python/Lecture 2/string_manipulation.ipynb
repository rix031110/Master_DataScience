```python
#Concatenation of str
s1="Hello world"
s2=" There is the sun"
s3=s1+s2
```

```python
#Slicing the str
s3[:10] #first ten characters

s3[-10:] #last ten character

s3[3:5] #the character in between excluding the closing one
```

```python
s3[2:9:2] #this add the number of steps used
```

```python
s3.upper() #returns the str in capital
```

```python
s3.find("sun") #it gives the number of the character in whioch the word start
```

```python
s3.replace("sun", "rain") #replace first par with the second
```

```python
s3.replace("sun", "rain".upper()) #combination of comands
```
