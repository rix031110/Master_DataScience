```python
try: # not to block the execution of the code if there is an error
    x=3/0  
except ZeroDivisionError as e:
    print(e) #if a division error occures, print generic error
```

```python
try:
    x = 10+"abc"
except TypeError as e:
    print(f"the following error has occurred: {e}")
```

```python
try:
    my_list=[1,2,3,4,5]
    print(my_list[10])

except Exception as e:   # to print any kind of errors
    print(f"the following error has occurred: {e}")
```
