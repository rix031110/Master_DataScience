# Lists

```python
my_list=[1,2,3,"Python", 4.5] #creating a list
my_list[4] #to select the 4th element of the lis (it starts from 0) 

print("List sample", my_list)
print("Fourth element of the list:",my_list[4]) 

my_list[1]= 20 #replace the element
my_list.append(True) #add an element at the end
print("List sample", my_list)

my_list.remove(my_list[3]) #remove an element
print("List sample", my_list)

my_list.insert(3,738939) #insert element at position 3
print("List sample", my_list)
```

# Dictionaries

```python
my_dictionary = {"name":"Alice", "age":25, "location":"New York"} # it is made by pairs of key and values
print(my_dictionary)

#Dictioanires are not ordere so we cannot select a specific one, as there is no first or secondo element
my_dictionary ["name"] #I can select based on keys

my_dictionary["country"]="United States" # add a pair of key and value
print(my_dictionary)

my_dictionary["tuple"]= my_tuple #you can add whatever you want in a dictionary
print(my_dictionary)

my_dictionary["tuple"][1] #to select an element from a tuple inside a dictionary
```
