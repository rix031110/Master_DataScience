```python
counter= 0
while counter<5:
    print(counter)
    counter+=1
"""
It control that the condition is matched, if it is it print the value of counter and modifies it adding 1 and restarting.
When the counter reaches 5, the condition is no more matched, so it will stop and not print counter = 5, so I will have the numebers from 0 to the
previous of the condition
"""
list=0
while list<5:
    print(list)
    list += 1
    if list==3:
        break
    
```
