```python
def is_even(a):
    result= a%2
    if result==1:
        print("It is odd")
    else:
        print("It is even")

is_even(10)
```

```python
def is_even2(a):
    if a%2==1:
        print("It is odd")
    else:
        print("It is even")

print(is_even2(10))
print(is_even2(11))
```

```python
def is_even3(a):
    if a%2==0:
        return True

    return False
print(is_even3(10))
print(is_even3(11))
```

```python
def is_even4(a):
    return (a%2==0) #as it is a boolean it naturally return True or False

print(is_even4(10))
print(is_even4(11))
```
