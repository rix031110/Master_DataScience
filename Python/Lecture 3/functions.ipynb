```python
def greet(name):
    name= name.upper()
    print(f"Hello {name}")
```

```python
greet("Alice")
```

```python
employee = "Salim"
greet(employee)
```

```python
def calculate_area(w, h):
    area= w*h
    return area

calculate_area(3,88)
```
