```python
"""
We can create desktop applications with Tkinter
"""
import tkinter as tk
dir(tk)
```

```python
root = tk.Tk()#iot creates the main window, in which we put all the widgets
```

```python
root.title("My First Tkinter window") #add title
```

```python
welcome_label=tk.Label(root, text="Welcome to Tkinter") #cretaes a labes in the main root
```

```python
welcome_label.pack() #it packages the label in the application
```

```python
def on_button_click(): #create a function to put in the button
    welcome_label.config(text="Eheheheh fregata") #let's change the text 
```

```python
click_me_button=tk.Button(root,text="Suka", command=on_button_click ) #create a button in root, with a text and which does a function
```

```python
click_me_button.pack()
```

```python
root.mainloop() #launces the desktop app
```
